# 로지스틱 회귀 - 분류

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Sigmoid 함수
- 만들어서 사용해야됨

```python
def sigmoid(x, a = 1, b = 0):
    return (1. /( 1 + np.exp(-a * (x - b))))
```

```python
xs = np.linspace(-5, 5, 1001)
ys = sigmoid(xs)
plt.plot(xs, ys, label = 'sigmoid')
plt.plot(xs, ys*(1-ys), label = 'der')
plt.title('Sigmoid function')
plt.yticks([0,0.5,1])
plt.grid()
plt.legend()
plt.show()
```

```python
y3 = sigmoid(xs, a = 3)
y_half = sigmoid(xs, a = 0.5)
plt.plot(xs, ys, label = 'sigmoid')
plt.plot(xs, y3, label = '3times')
plt.plot(xs, y_half, label = 'half')
plt.title('Sigmoid function')
plt.yticks([0,0.5,1])
plt.grid()
plt.legend()
plt.show()
```

## 이진분류

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer.data.shape
```

- 로지스틱 회귀를 사용하기 위한 정규화 or 표준화

```python
from sklearn.preprocessing import StandardScaler
cancer_std = StandardScaler().fit_transform(cancer.data)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    cancer_std, cancer.target, stratify = cancer.target, random_state = 2022)
```

```python
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression(random_state= 2022)
lrc.fit(X_train, y_train)
```

```python
lrc.coef_
```

```python
lrc.intercept_
```

```python
X_test[0]
```

```python
# np.dot  : 행렬 곱
val = np.dot(lrc.coef_, X_test[0]) + lrc.intercept_
val
```

```python
sigmoid(val)
```

```python
lrc.predict_proba(X_test[:5])
```

## 다중분류

```python
from sklearn.datasets import load_wine
wine = load_wine()
wine_std = StandardScaler().fit_transform(wine.data)
X_train, X_test, y_train, y_test = train_test_split(
    wine_std, wine.target, stratify = wine.target, random_state = 2022)
```

```python
lrc2 = LogisticRegression(random_state=2022)
lrc2.fit(X_train, y_train)
```

```python
lrc2.predict_proba(X_test[:5])
```

```python
lrc2.predict(X_test[:5])
```

```python
lrc2.coef_
```

```python
lrc2.intercept_
```

```python
np.dot(lrc2.coef_,X_test[0])
```

```python
val2 = np.dot(lrc2.coef_,X_test[0]) + lrc2.intercept_
val2
```

```python
sigmoid(val2)
```
